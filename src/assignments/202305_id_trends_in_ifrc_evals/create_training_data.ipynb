```python
def write_text(file_content, pdf):
    
    
    filename = pdf.replace(".pdf",".txt")
    full_filepath = os.path.join(corpora_location,'evals_in_text',filename)
    
    
    with open(full_filepath, "w", encoding="utf-8") as f:
        f.write(file_content)
        
    
```

```python
# get available evaluation files
#pdf_list = ['2022_pdm_report_of_the_mdrnp011_nepal_monsoon_floods_and_landslides.pdf']
pdf_list = [pdf for pdf in os.listdir(corpora_location) if pdf[-4:] == '.pdf']

#df_paragraph = pd.DataFrame(columns=['file','paragraph'])
text_content = ""

i=0
for pdf in pdf_list:
    i+=1
    print(i)
    filepath = os.path.join(corpora_location,pdf)
    
    try:
        #df_paragraph = pd.concat([df_paragraph, parse_pdf.load_pdf_to_df(filepath)], ignore_index=True)
        file_content = parse_pdf.read_pdf_document(filepath)
        write_text(file_content, pdf)

        text_content = text_content + "\n" + file_content
    except Exception as e:
        print(e)
        pass
    
```

## iterate over files and pre-process them

```python
txt_list = [txt for txt in os.listdir(corpora_location) if txt[-4:] == '.txt']

#df_paragraph = pd.DataFrame(columns=['file','paragraph'])
text_content = ""
df = pd.DataFrame(columns=['file','original_text','preprocessed_text'])


i=0
segments=[]
for txt in txt_list:
    i+=1
    print(f"file num {i} : df length is {df.shape[0]}")
    filepath = os.path.join(corpora_location,txt)
    
    try:
        #df_paragraph = pd.concat([df_paragraph, parse_pdf.load_pdf_to_df(filepath)], ignore_index=True)
        with open(filepath,"r", encoding="utf-8") as f:
            file_content = f.read()

            
            doc = nlp(file_content)
            segments = []
            #df = pd.DataFrame(columns=['file','original_text'])
            
    
            for sent in doc.sents:
                row = [txt]

                sent_after_stops = ' '.join([str(w.lemma_) for w in sent if w.is_stop == False])
                sent_after_stops = sent_after_stops.replace("\n"," ")
                #segment = sent.text.replace("\n"," ")
                segment = simple_preprocess(sent_after_stops, deacc=True) #<--do this later
                #segment = sent_after_stops
                if len(segment) > 1 :
                    row.append(sent.text)
                    row.append(segment)
                    #segments.append(segment)

                    df.loc[len(df.index)] = row 
            
                    #df = pd.concat(df, row, ignore_index=True)

    except Exception as e:
        print(e)
        pass
    
    
df.to_pickle("eval_sents.pkl")
```
